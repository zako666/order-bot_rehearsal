```python
#總收入num,總相異顧客人num2,總相異商品數num3,總相異交易時間數num4
#每種商品賣出幾個num5
#統計各顧客買過什麼有幾個num6
#統計顧客最常買的商品num7
#顧客ID,購買時間,購買商品,購買數量,商品小計,
f=open('text_value.csv',"r",encoding="Big5")
d=[]
num,num2,num3,num4,num5,num6,num7=0,set(),set(),set(),{},[],{}
for l in f:
    t=l.split(",")
    d.append(t)
for i in range(1,len(d)):
    num+=int(d[i][4])  #總收入num
    num2.add(d[i][0])  #總相異顧客人num2
    num3.add(d[i][2])  #總相異商品數num3
    num4.add(d[i][1])  #總相異交易時間數num4
k=list(num3)
for i in k:    #歸零個商品交易數
    num5[i]=int(0)
for i in range(1,len(d)): #sum各商品交易數
    g=0
    for j in k:
        if j==d[i][2]:
            g=d[i][2]
    l=num5.pop(g)
    if g!=0:
        num5[g] = l +int(d[i][3])
    g=0
    
a=0
for i in range(len(num2)): #顧客名單輸入num6
    num6.append([list(num2)[i]])
    for j in range(len(num3)): #商品輸入
        num6[a].append([list(num3)[j],0])
    a+=1 

for i in range(1,len(d)): #統計各顧客買過什麼有幾個
    h=0
    for j in range(len(num2)):
        if d[i][0]==list(num2)[j]:
            m=1
            for k in range(len(num3)):
                 if d[i][2]==list(num3)[k] :
                     num6[h][m][1]+=1
                 m=(m+1)%len(num3)
                 if m==0:
                    m=1
        h=(h+1)%len(num2)


c=0
b=list(num2)
for i in b:    #歸零顧客最常買的商品名
    num7[i]=''
for i in range(len(num2)):
    a=0
    for j in range(1,len(num3)):
         if num6[i][j][1]>a:
                a=num6[i][j][1]
                num7[list(num2)[i]]=num6[i][j][0] 
       
print("總收入: ",num)
print("客人: ",num2," ",len(num2))
print("商品: ",num3," ",len(num3))
print("交易時間: ",num4," ",len(num4))
print("各商品交易量: ",num5.items())
# print("各商品交易量: ",num6)
print("各顧客最常買的商品: ",num7.items())
f.close()
```
