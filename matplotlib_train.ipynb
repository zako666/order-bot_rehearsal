```python
#垂直長條圖 plt.bar(index直的, value)
#X軸index替換 plt.xticks(index, labels)
labels = ["Python", "C++", "Java", "JS", "C", "C#"]
index = np.arange(len(labels))
ratings = [5.168,5.726,14.988,3.165,11.857,4.453]
plt.bar(index, ratings)
plt.xticks(index, labels)
plt.ylabel("Usage")
plt.title("Programming Language Usage")
plt.show()

#水平長條圖 plt.barh(index直的, value)
plt.barh(index, ratings)
plt.yticks(index, labels)
plt.xlabel("Usage")
plt.title("Programming Language Usage")
plt.show()
```

```python
#派圖(圓餅圖) plt.pie(ratings, labels=labels)
labels = ["Python", "C++", "Java", "JS", "C", "C#"]
ratings = [5, 6, 15, 3, 12, 4]
plt.pie(ratings, labels=labels)
plt.title("Programming Language Usage")
plt.axis("equal")
plt.show()

#使用突增值 plt.pie(ratings, labels=labels,explode=explode)
#色彩圖例 plt.legend(patches, labels, loc="best")
explode = (0, 0,0, 0.2,0, 0.2)
plt.pie(ratings, labels=labels,explode=explode)
plt.legend(patches, labels, loc="best")
plt.show()
```
